```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\user\Desktop\Python\DAP exam\climate_change_data.csv")
```

```python
data
```

```python
data['Date'].value_counts()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(100, 50))
```

```python
plt.hist(data['Temperature'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Temperature')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
```

```python
plt.hist(data['CO2 Emissions'], bins=20, color='skyblue', edgecolor='black')
plt.title('CO2 Emissions')
plt.xlabel('Date')
plt.ylabel('CO2 Emissions (ppm)')
```

```python
plt.hist(data['Sea Level Rise'], bins=20, color='skyblue', edgecolor='black')
plt.title('Sea Level Rise')
plt.xlabel('Date')
plt.ylabel('Sea Level Rise (mm)')
```

```python
plt.hist(data['Precipitation'], bins=20, color='skyblue', edgecolor='black')
plt.title('Precipitation')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
```

```python
plt.hist(data['Humidity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Humidity')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
```

```python
plt.hist(data['Wind Speed'], bins=20, color='skyblue', edgecolor='black')
plt.title('Wind Speed')
plt.xlabel('Date')
plt.ylabel('Wind Speed (m/s)')
```

```python
plt.tight_layout()
```
